```python
import pandas as pd
```

```python
death_data = pd.read_csv('./raw/全死因1110822/dead110.txt', dtype={'county':str})
```

```python
death_data
```

```python
county_code = pd.read_csv('./raw/全死因1110822/processed/county_code.csv', sep='\t', dtype={'county':str})
```

```python
county_code
```

```python
sex_code = pd.read_csv('./raw/全死因1110822/processed/sex.csv', sep='\t')
sex_code
```

```python
age_code = pd.read_csv('./raw/全死因1110822/processed/age_code.csv', sep='\t')
age_code
```

```python
cause_code = pd.read_csv('./raw/全死因1110822/processed/cause.csv', sep='\t')
cause_code
```

```python
death_data_joined = death_data.join(county_code.set_index('county'), on='county')\
                                .join(sex_code.set_index('sex'), on='sex')\
                                .join(age_code.set_index('age_code'), on='age_code')\
                                .join(cause_code.set_index('109年以後cause'), on='cause')
```

```python
death_data_joined
```

```python
list(death_data_joined[death_data_joined['100年~鄉鎮市區'].str.contains('桃園縣')]['100年~鄉鎮市區'].unique())
```

```python
def replace_words(x):
    if '桃園縣' in x:
        y = list(x)
        y[2] = '市'
        y[5] = '區'
    else:
        y = x
    return ''.join(y)

death_data_joined['100年~鄉鎮市區'] = death_data_joined['100年~鄉鎮市區'].apply(lambda x: replace_words(x))
```

```python
death_data_joined
```

```python
list(death_data_joined[death_data_joined['100年~鄉鎮市區'].str.contains('桃園市')]['100年~鄉鎮市區'].unique())
```

```python
death_data_joined.to_csv('./data/dead110_joined.csv', index=False)
```
